## import dependencies

```java
import tech.tablesaw.api.*;
import static tech.tablesaw.aggregate.AggregateFunctions.*;
import tech.tablesaw.aggregate.NumericAggregateFunction;
import tech.tablesaw.plotly.*;
import tech.tablesaw.plotly.api.AreaPlot;
import tech.tablesaw.plotly.components.*;

StringColumn nc = StringColumn.create("name", new String[] {"A", "A", "B", "B"});
DoubleColumn nc2 = DoubleColumn.create("score",  new double[] {1, 2, 3, 4});
DoubleColumn nc3 = DoubleColumn.create("score2",  new double[] {5, 3, 3, 4});

System.out.println(nc);
System.out.println(nc.where(nc.isEqualTo("A")).print());
System.out.println(Table.create().addColumns(nc, nc2).print(2));
Table df = Table.create().addColumns(nc, nc2, nc3);
Table df1 = Table.create().addColumns(nc, nc2, nc3);
Table df2 = Table.create().addColumns(nc2);
Table df3 = Table.create().addColumns(nc3);
display(nc2.where(nc2.isIn(nc3.asDoubleArray())).print());
display(df2.concat(df3));
display(df.append(df1));
df.summarize("score", mean, sum, min, max).by(nc);
NumericAggregateFunction ff = new NumericAggregateFunction("ff") {
    public Double summarize(NumericColumn<?> column) {
        return Double.valueOf(column.getDouble(0));
    }
};
df.summarize("score","score2", ff, max).by(nc);
```
